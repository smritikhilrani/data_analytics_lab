## Standardising the data

```python
file3.rename(columns={"State" : "ST"}, inplace=True)
```

```python
file3.rename(columns={"Gender" : "GENDER"}, inplace=True)
```

```python
def concat_data():

    return pd.concat([file1,file2,file3], axis=0)
#concatinating the three files
```

```python
cd_df = concat_data()
```

```python
cd_df
```

```python
# Dropping duplicate values in the dataframe
cd_df1 = cd_df.drop_duplicates()
```

```python
#Drop column Customer since it is a unique value which identifies the row.
cd_df1.drop(['Customer'], axis=1, inplace= True)
```

```python
cd_df1.info() #checking the type of all the variables
```

```python
cd_df1.head()
```

### Fixing incorrect values in the data
#### remove the percentage from column 'customer lifetime value' and truncate the values.

```python
cd_df1["Customer Lifetime Value"] = cd_df1["Customer Lifetime Value"].apply(lambda x: float(x.rstrip('%')) / 100.0  if isinstance(x,str) else x)
```

```python
cd_df1
```

```python
# Truncate
cd_df1 = cd_df1.round({'Customer Lifetime Value': 0})
```

```python
cd_df1
```

### Correcting the column with values : 1/0/00 format in column Number of Open Complaints

```python
#separating values with '/' and solitting into three columns
cd_df1[['0','1','2']] = cd_df1['Number of Open Complaints'].str.split(pat='/', expand=True)
cd_df1
```

```python
print(cd_df1[cd_df1['Number of Open Complaints'].isnull()]) #checking for nulls in the column
```

```python
cd_df1.dropna(subset=['Number of Open Complaints'],inplace = True) #removing the null value
```

```python
#extracting the middle value 
cd_df1['Number of Open Complaints'] = cd_df1['1'].combine_first(cd_df1['Number of Open Complaints']).astype('int64')
cd_df1
```

```python
cd_df1['Number of Open Complaints'].value_counts() #checking the values
```

### Standardising Gender and State values

```python
cd_df1['GENDER'].value_counts()
```

```python
def gender_value(GENDER):
    if GENDER == 'Male' or GENDER == 'M':
        return 'M'
    elif GENDER == 'Female' or GENDER == 'Femal' or GENDER == 'F':
        return 'F'
    
cd_df1['GENDER2'] = cd_df1.GENDER.apply(gender_value)   
```

# Tuesday

#### replacing null values with its respective column's mean

```python
cd_df1["Total Claim Amount"] = cd_df1["Total Claim Amount"].fillna(value = cd_df1[['Total Claim Amount']].mean())
```

```python
mean_ltv = cd_df1['Customer Lifetime Value'].mean()
```

```python
mean_mp = cd_df1['Monthly Premium Auto'].mean()
```

```python
mean_tc = cd_df1['Total Claim Amount'].mean()
```

```python
values = {"Customer Lifetime Value": mean_ltv, "Monthly Premium Auto": mean_mp, "Total Claim Amount": mean_tc}
cd_df1 = cd_df1.fillna(value=values)
```

```python
cd_df1.info()
```

```python
print(cd_df1[cd_df1['Customer Lifetime Value'].isna()])
```

#### replace zeros from income

```python
cd_df2 = cd_df1.copy()
```

```python
mean_value = cd_df2['Income'].mean()
```

```python
cd_df2['Income']=round(cd_df2.Income.replace(0,mean_value),2)
```

## Grouping state by zones

```python
Zones = {'California' : 'West', 'Oregon' : 'North West', 
               'Washington' : 'East', 'Arizona' : 'Central', 'Nevada' : 'Central'}
cd_df3 = cd_df2.assign(Groups=cd_df2.ST.map(Zones)).fillna('Other')  
```
