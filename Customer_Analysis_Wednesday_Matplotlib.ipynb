#### find the total number of responses

```python
response_rate = categorical.response.value_counts()
```

```python
fig2,ax = plt.subplots(1,1,figsize = (8,5))
plt.title('Total Number of Responses', fontsize=18)
ax.bar(response_rate.index, response_rate , color='green', label='Responses')
#ax[0].set_title('Total Number of Responses', fontsize=15)
```

##### *using another method to plot total responses*

```python
ax = categorical['response'].value_counts()[:20].plot(kind='barh')
plt.legend(loc = 'upper left')
```

## Response by Sales Channel

```python
# grouping by sales channel
response_per_sales_channel = categorical.groupby(['sales_channel','response'])['sales_channel'].count().rename('count').reset_index()
```

```python
response_per_sales_channel
```

```python
sns.set_palette('RdYlGn')
palette_response = {"yes": "yellow", "no": "red"}
```

```python
# Plotting the graph
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
plt.title('Response count by sales channel', size='15', fontweight='bold')

sns.barplot(x = response_per_sales_channel['sales_channel'],y = response_per_sales_channel['count'], hue = response_per_sales_channel['response'], palette=palette_response)
```

### Response by Total Claim amount

```python
# Binning the total claim amount
df['new_total_claim_amount'] = pd.cut(x=df['total_claim_amount'], bins=[0,500, 1000, 1500, 2000, 2500, 3000],
                    labels=['500', '1000', '1500',
                            '2000', '2500', '3000'])
```

```python
# Grouping by total claim amount
response_by_total_claim_amount = df.groupby(['new_total_claim_amount','response'])['new_total_claim_amount'].count().rename('count').reset_index()
```

```python
response_by_total_claim_amount
```

```python
# Plotting a bar graph
sns.barplot(x = response_by_total_claim_amount['new_total_claim_amount'],y = response_by_total_claim_amount['count'], hue = response_by_total_claim_amount['response'], palette=palette_response)
```

## Response by Income

```python
# Setting the palette colors for x and y axis
sns.set_palette('RdYlGn')
palette_response = {"yes": "green", "no": "blue"}
```

```python
# Binning the income
df['new_income'] = pd.cut(x=df['income'], bins=[0,20000, 40000, 60000, 80000, 100000],
                    labels=['20000', '40000', '60000',
                            '80000', '100000'])
```

```python
df
```

```python
# Grouping by income
response_by_income = df.groupby(['new_income','response'])['new_income'].count().rename('count').reset_index()
```

```python
response_by_income
```

```python
sns.barplot(x = response_by_income['new_income'],y = response_by_income['count'], hue = response_by_income['response'], palette=palette_response)
```
